```python
# Face detection using Haar Cascade Features
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

single=cv2.imread('single.jpg',0)
group=cv2.imread('group.jpg',0)

face_casscade=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
eye_cas=cv2.CascadeClassifier('haarcascade_eye.xml')

def detect_face(img):
    face_img=img.copy()
    face_rects=face_casscade.detectMultiScale(face_img)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(0,255,0),10)
    return face_img

def adj_detect_face(img):
    face_img=img.copy()
    face_rects=face_casscade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in face_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img




result_face=detect_face(single)
plt.imshow(result_face,cmap='gray')
result_adj_face=adj_detect_face(group)
plt.imshow(result_adj_face,cmap='gray')
```

```python
# Eyes detection using Haar Cascade Features
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

single=cv2.imread('single.jpg',0)
eye_cas=cv2.CascadeClassifier('haarcascade_eye.xml')

def detect_eyes(img):
    face_img=img.copy()
    eye_rects=eye_cas.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    for (x,y,w,h) in eye_rects:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,255,255),10)
    return face_img

result=detect_eyes(single)
plt.imshow(result,cmap='gray')
```

```python
# Live Face detection using Haar Cascade Features
```

```python
cap=cv2.VideoCapture(0)

while True:
    ret,frame=cap.read(0)
    frame=detect_face(frame)
    cv2.imshow('Video',frame)
    
    if cv2.waitKey(1) & 0xFF ==ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
